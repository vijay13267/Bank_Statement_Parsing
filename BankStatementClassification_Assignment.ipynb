# 1.Reading data From a "Bank Statement" PDF file

```python
import pandas as pd
```

```python
from tabula import read_pdf
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
data=read_pdf('./Assignment.pdf',pages='all',lattice=True,pandas_options={'header':None},)
```

```python
data
```

```python
df1=data[0].iloc[1:]
```

```python
df2=data[2].iloc[1:]
```

```python
frames=pd.concat([df1,df2])
```

```python
frames
```

# 2. Extracting Required Data

```python
frames.index=[i for i in range((frames.shape)[0])]
```

```python
frames
```

```python
main={}
del_rows=[]
for i in range(1,(frames.shape)[0]+1):
    if (frames.iloc[i:i+1,0:1]).isnull().values.any():
        x=str(frames.iloc[i:i+1,1:2])
        main[x]=frames.iloc[i:i+1]
        del_rows.append(i)
del_rows
```

## 2.1. Actual Transaction Data After Removing unnecessary rows

```python
actual_data=frames.drop(del_rows,axis=0)
```

```python
actual_data
```

#2.2. Only Credits Transactions

```python
credits_data=actual_data[actual_data[3].isna()].reset_index()
#print(credits_data.shape)
credits_data
```

# 2.3. Only Debits Transactions

```python
debits_data=actual_data[actual_data[4].isna()].reset_index()
#print(debits_data.shape)
debits_data
```

# 3. Categorization

## 3.1. Analysing the savings of that Month

```python
debits=actual_data[3][1:].apply(pd.to_numeric, errors='ignore')
debits=debits[debits>0]
debits=pd.Series(debits)
print('Total Debits: ',debits.sum())
debits.describe()
```

```python
credits=actual_data[4][1:].apply(pd.to_numeric, errors='ignore')
credits=credits.values.tolist()
credits=pd.Series(credits)
print('Total credits: ',credits.sum())
credits.describe()
```

```python
monthly_analysis=credits.sum()-debits.sum()
monthly_analysis
```

- <b>conclusion</b>:

     - As the monthly_analysis at the end of the month is negative, The person had spent more than that of the money he earned  during the month
    
     - Based on the monthly Analysis, We can get information such as the savings of the person, his growth(by comparing previous month statements), his living standards etc
     
     <b>Here,negative monthly analysis indicates he/she using money that was saved in the last month</b>
     

## 3.2. Representation of Monthly Analysis(credit, debit) 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.pie([credits.sum(),debits.sum()],labels=['credits','debits'],autopct="%0.2f")
plt.title('Credits Vs Debits')
plt.show()
```

## 3.3. Categorizing transactions('IMPS','NEFT','Cash Deposit') 

```python
import re

categories=dict((i,0) for i in ['IMPS','NEFT','CASH DEPOSIT'])

particulars=actual_data[2].values.tolist()
particulars="".join(particulars)
#print(particulars)
for i in categories.keys():
    x=re.findall(i,particulars)
    categories[i]=len(x)
categories['others']=len(actual_data[0])-sum(categories.values())
print('The Number of Transaction in each category :',categories)
```

## 3.4. Representing Transactions categories

```python
plt.pie(categories.values(),labels=categories.keys(),autopct="%0.2f", )
plt.show()
```

# 4. Final Results

```python
print('Total Credits\t:',credits.sum())
print('Total Debits\t:',debits.sum())

print('Highest Credit\t:',credits.max())
print('Highest Debit\t:',debits.max())

print('\nSpent More than that of Earned\t:',credits.sum()<debits.sum())

print('\nTypes and frequency of Transactions\t:',categories)
```
